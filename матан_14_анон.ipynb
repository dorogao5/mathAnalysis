1. Программа которая по заданной формуле членов последовательности рисует
набор из N точек вида $(k, x_{k})$, где $k = [1, N] ∩ \mathbb{N}$.

P.S. На данном и всех последующих графиках, может быть неочевидно, но значениями последовательности являются только точки. Линия на графике лишь лучшей видимости того, как идёт последовательность.

```python
import numpy as np
from matplotlib import pyplot as plt
def f(x):
    term1 = (3 - (-1) ** x)
    term2 = (2 * x + 5) / (x + 2)
    return term1 * term2
x = np.array(list(range(1, 101)))
y = f(x)
plt.figure(figsize=(12, 6), facecolor='#1c092b')
ax = plt.gca()
ax.set_facecolor('#11041e')
plt.scatter(x, y, color='#8020a2', label='Значения последовательности', edgecolor='black', alpha=0.7)
plt.plot(x, y, color='#8020a2', label='Направление линии последовательности', alpha=0.7)
plt.scatter(25, f(25), color='#d580ff', s=52, edgecolor='white', label='$x_{n}$ некая выделенная точка (для примера)')
plt.axhline(y=0, color='#5c238e', linestyle='--', linewidth=0.8, alpha=0.6)
plt.title('Визуализация последовательности $x_{n}= (3-(-1)^n)\\cdot\\frac{2n+5}{n+2}$', fontsize=14, color='#d3b2ff')
plt.xlabel('Значения $n \in \mathbb{N}$', fontsize=15, color='#d3b2ff')
plt.ylabel('Значения $x_{n}$', fontsize=12, color='#d3b2ff')
plt.grid(visible=True, which='both', color='#5c238e', linestyle='--', linewidth=0.7, alpha=0.5)
plt.xticks(np.arange(0, 110, 10), color='#d3b2ff')  # Шаг значений по оси x в светло-пурпурном
plt.yticks(np.linspace(min(y) - 2, max(y) + 3, 5), color='#d3b2ff')  # Шаг значений по оси y в светло-пурпурном
legend = plt.legend(facecolor='#1c092b')
for text in legend.get_texts():
    text.set_color('#d3b2ff')
plt.show()
```

4-5. По данному $\varepsilon>0$ определим номер подпоследовательности $k_{0}$, начиная с которого, элементы подпоследовательности будут попадать в $\varepsilon \text{-окрестность}$ предела. Отедельно выделим те точки, которые попадают в данную $\varepsilon \text{-окрестность}$ и отметьте горизонтальной линией значение предела этой подпоследовательности.


1) Пусть $ɛ = 0.1$:

```python
import numpy as np
from matplotlib import pyplot as plt
def f(x):
    term1 = (3 - (-1) ** x)
    term2 = (2 * x + 5) / (x + 2)
    return term1 * term2
x = np.array(list(range(1, 101)))
y = f(x)
k = np.array(list(range(0, 50)))
x_subsequence = 2 * (4 * k + 5) / (2 * k + 2)
x_indices = 2 * k
epsilon = 0.1
limit = 4
indices_in_epsilon = np.where(np.abs(x_subsequence - limit) < epsilon)[0]
if len(indices_in_epsilon) > 0:
    k_0 = np.min(indices_in_epsilon)
    x_in_epsilon = x_indices[indices_in_epsilon]
    y_in_epsilon = x_subsequence[indices_in_epsilon]
    print(k_0)
plt.figure(figsize=(12, 6), facecolor='#1c092b')
ax = plt.gca()
ax.set_facecolor('#11041e')
plt.scatter(x, y, color='#8020a2', label='Значения последовательности $x_n$', edgecolor='black', alpha=0.7)
plt.plot(x, y, color='#8020a2', alpha=0.7)
plt.scatter(x_indices, x_subsequence, color='red', label='Сходящаяся подпоследовательность $x_{2k+1}$', edgecolor='black', alpha=0.8)
plt.axhline(y=limit, color='cyan', linestyle='--', linewidth=1.2, label=f'Предел подпоследовательности = {limit}')
if len(x_in_epsilon) > 0:
    plt.scatter(x_in_epsilon, y_in_epsilon, color='orange', label=f'Точки в $\\varepsilon$-окрестности ($\\varepsilon={epsilon}$)', s=60, edgecolor='black', alpha=0.9)
plt.title(f'Окрестность предела подпоследовательности ($\\varepsilon={epsilon}$)', fontsize=14, color='#d3b2ff')
plt.xlabel('Значения $n \\in \\mathbb{N}$', fontsize=15, color='#d3b2ff')
plt.ylabel('Значения $x_{n}$', fontsize=12, color='#d3b2ff')
plt.grid(visible=True, which='both', color='#5c238e', linestyle='--', linewidth=0.7, alpha=0.5)
plt.xticks(np.arange(0, 110, 10), color='#d3b2ff')
plt.yticks(np.linspace(min(y) - 2, max(y) + 3, 5), color='#d3b2ff')
legend = plt.legend(facecolor='#1c092b')
for text in legend.get_texts():
    text.set_color('#d3b2ff')
plt.show()
```

k0 = 9 — первый индекс, начиная с которого элементы подпоследовательности попадают в ε-окрестность.

При заданном ε > 0 необходимо ищем такие элементы $x_{m}$ последовательности, чтобы выполнялось неравенство $x_{m}< inf\text{ }x_{n}+\varepsilon$ и выделяем их на графике, при необходимости изменяя масштаб и число отображаемых точек.
