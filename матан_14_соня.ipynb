1. Программа которая по заданной формуле членов последовательности рисует
набор из N точек вида $(k, x_{k})$, где $k = [1, N] ∩ \mathbb{N}$.


```python
# Импорт необходимых библиотек
import numpy as np
from matplotlib import pyplot as plt

# Определяем функцию f для последовательности
def f(x):
    return (2 + np.cos(np.pi * x / 2)) * (3 * x - 1) / (x + 2)

# Генерируем массив натуральных чисел от 1 до 100
x = np.array(list(range(1, 101)))
y = f(x)

# Создаем график
plt.figure(figsize=(12, 6))
plt.style.use('Solarize_Light2')  # Установка светлого стиля

plt.scatter(x, y, color='#000000', label='Значения последовательности', edgecolor='white', alpha=0.7)  # Цвет точек
plt.scatter(25, f(25), color='#ff5733', s=52, edgecolor='black', label='$x_{n}$ точка (для примера)')  # Подсветка некой точки

# Плавная кривая для общего поведения последовательности
x_smooth = np.linspace(1, 100, 500)
y_smooth = f(x_smooth)
plt.plot(x_smooth, y_smooth, color='#49423d', alpha=0.7, linestyle='-', linewidth=2, label='Сглаженная кривая')

# Горизонтальная линия для справки
plt.axhline(y=0, color='darkgrey', linestyle='--', linewidth=0.8, alpha=0.6)

# Заголовки и метки осей
plt.title('Визуализация последовательности $x_{n} = \\left( 2 + \\cos \\frac{\\pi n}{2} \\right) \\cdot \\left( \\frac{3n - 1}{n + 2} \\right)$', fontsize=14, color='#49423d')
plt.xlabel('Значения $n \in \mathbb{N}$', fontsize=15, color='#49423d')
plt.ylabel('Значения $x_{n}$', fontsize=12, color='#49423d')

# Добавление сетки
plt.grid(visible=True, which='both', color='grey', linestyle='--', linewidth=0.7, alpha=0.5)

# Показ значений на осях
plt.xticks(np.arange(0, 110, 10), color='#49423d')  # Шаг значений по оси x
plt.yticks(np.linspace(min(y) - 5, max(y) + 5, 15), color='#49423d')  # Шаг значений по оси y

# Легенда с изменением цвета текста
legend = plt.legend()
for text in legend.get_texts():
    text.set_color('black')  # Установка цвета текста легенды на черный

# Показ графика
plt.show()
```

3. Выделим сходящуюся последовательность $x_{4k}=\frac{12k-1}{4k+2}$

```python
# Импорт необходимых библиотек
import numpy as np
from matplotlib import pyplot as plt

# Определяем функцию f для последовательности
def f(x):
    return (2 + np.cos(np.pi * x / 2)) * (3 * x - 1) / (x + 2)

# Определяем сходящуюся подпоследовательность
def g(k):
    return 3*(12 * k - 1) / (4 * k + 2)

# Генерируем массив натуральных чисел от 1 до 100
x = np.array(list(range(1, 101)))
y = f(x)

# Генерируем массив для подпоследовательности x_{4k}
k = np.array(list(range(1, 26)))  # k от 1 до 25 для x_{4k}, чтобы покрыть диапазон n до 100
x_4k = 4 * k
y_4k = g(k)

# Создаем график
plt.figure(figsize=(12, 6))
plt.style.use('Solarize_Light2')  # Установка светлого стиля

plt.scatter(x, y, color='#000000', label='Значения последовательности $x_{n}$', edgecolor='white', alpha=0.7)  # Основные точки

# Плавная кривая для общего поведения последовательности
x_smooth = np.linspace(1, 100, 500)
y_smooth = f(x_smooth)
plt.plot(x_smooth, y_smooth, color='#49423d', alpha=0.7, linestyle='-', linewidth=2, label='Сглаженная кривая')

# Горизонтальные линии для заданных значений
plt.axhline(y=9, color='blue', linestyle='--', linewidth=1.5, label='$\\sup x_{n} = 9, \\overline{\\lim}_{n \\to \\infty} x_{n} = 9$')

# Выделение подпоследовательности x_{4k}
plt.scatter(x_4k, y_4k, color='purple', s=70, edgecolor='white', label='Подпоследовательность $x_{4k} = 3 \cdot \\frac{12k - 1}{4k + 2}$')

# Заголовки и метки осей
plt.title('Визуализация последовательности $x_{n} = \\left( 2 + \\cos \\frac{\\pi n}{2} \\right) \\cdot \\left( \\frac{3n - 1}{n + 2} \\right)$ и подпоследовательности $x_{4k}$', fontsize=14, color='#49423d')
plt.xlabel('Значения $n \\in \\mathbb{N}$', fontsize=15, color='#49423d')
plt.ylabel('Значения $x_{n}$', fontsize=12, color='#49423d')

# Добавление сетки
plt.grid(visible=True, which='both', color='grey', linestyle='--', linewidth=0.7, alpha=0.5)

# Показ значений на осях
plt.xticks(np.arange(0, 110, 10), color='#49423d')  # Шаг значений по оси x
plt.yticks(np.linspace(min(y) - 5, max(y) + 5, 15), color='#49423d')  # Шаг значений по оси y

# Легенда с изменением цвета текста
legend = plt.legend()
for text in legend.get_texts():
    text.set_color('black')  # Установка цвета текста легенды на черный

# Показ графика
plt.show()
```

4-5. По данному $\varepsilon>0$ определим номер $n_{0}$, начиная с которого, элементы подпоследовательности будут попадать в $\varepsilon \text{-окрестность}$ предела. Отедельно выделим те точки, которые попадают в данную $\varepsilon \text{-окрестность}$ и отметьте горизонтальной линией значение предела этой подпоследовательности.


1) Пусть $ɛ = 0.1$:

```python
# Импорт необходимых библиотек
import numpy as np
from matplotlib import pyplot as plt

num_x1 = 0
num_x2 = 220

# Определяем функцию f для последовательности
def f(x):
    return (2 + np.cos(np.pi * x / 2)) * (3 * x - 1) / (x + 2)

# Определяем правильную сходящуюся подпоследовательность
def g(k):
    return 3 * (12 * k - 1) / (4 * k + 2)

# Генерируем массив натуральных чисел от 400 до 800
x = np.array(list(range(num_x1, num_x2)))
y = f(x)

# Генерируем массив для подпоследовательности x_{4k} в диапазоне n = 400 до 800
k = np.array(list(range(num_x1//4, num_x2//4)))  # k от 100 до 200, чтобы 4k было в диапазоне [400, 800]
x_4k = 4 * k
y_4k = g(k)

# Предел подпоследовательности
limit = 9
epsilon = 0.1

# Находим точки, которые попадают в epsilon-окрестность предела
within_epsilon_indices = np.where(np.abs(y_4k - limit) < epsilon)[0]
x_4k_within_epsilon = x_4k[within_epsilon_indices]
y_4k_within_epsilon = y_4k[within_epsilon_indices]

# Определяем n0 - первый индекс, начиная с которого все элементы подпоследовательности находятся в окрестности
n0 = x_4k_within_epsilon[0] if len(x_4k_within_epsilon) > 0 else None

# Создаем график
plt.figure(figsize=(12, 6))
plt.style.use('Solarize_Light2')  # Установка светлого стиля

plt.scatter(x, y, color='#000000', label='Значения последовательности $x_{n}$', edgecolor='white', alpha=0.7)  # Основные точки
plt.scatter(x_4k, y_4k, color='purple', s=70, edgecolor='white', label='Подпоследовательность $x_{4k} = 3 \\cdot \\frac{12k - 1}{4k + 2}$')  # Подпоследовательность
plt.scatter(x_4k_within_epsilon, y_4k_within_epsilon, color='red', s=100, edgecolor='black', label='Точки в $\\varepsilon$-окрестности предела')  # Точки в epsilon-окрестности

# Горизонтальная линия для предела
plt.axhline(y=limit, color='green', linestyle='--', linewidth=1.5, label='Предел подпоследовательности $x_{4k} = 9$')

# Заголовки и метки осей
plt.title('Визуализация подпоследовательности $x_{4k}$ и $\\varepsilon$-окрестности предела при epsilon = 0.001', fontsize=14, color='#49423d')
plt.xlabel('Значения $n \\in \\mathbb{N}$', fontsize=15, color='#49423d')
plt.ylabel('Значения $x_{n}$', fontsize=12, color='#49423d')

# Добавление сетки
plt.grid(visible=True, which='both', color='grey', linestyle='--', linewidth=0.7, alpha=0.5)

# Показ значений на осях
plt.xticks(np.arange(num_x1, num_x2, 100), color='#49423d')  # Шаг значений по оси x
plt.yticks(np.linspace(min(y) - 1, max(y) + 1, 15), color='#49423d')  # Шаг значений по оси y

# Легенда с изменением цвета текста
legend = plt.legend()
for text in legend.get_texts():
    text.set_color('black')  # Установка цвета текста легенды на черный

# Показ графика
plt.show()

# Выводим n0
if n0:
    print(f"Начиная с n0 = {n0}, элементы подпоследовательности попадают в epsilon-окрестность предела.")
else:
    print("Нет элементов подпоследовательности, попадающих в указанную epsilon-окрестность.")
```

Начиная с n0 = 208, элементы подпоследовательности попадают в эписолн-окрестность предела.
