1. Программа которая по заданной формуле членов последовательности рисует
набор из N точек вида $(k, x_{k})$, где $k = [1, N] ∩ \mathbb{N}$.


```python
# Импорт необходимых библиотек
import numpy as np
from matplotlib import pyplot as plt

# арккотангенс
def arccot(x):
    with np.errstate(divide='ignore', invalid='ignore'):
        result = np.where(x != 0, np.arctan(1 / x), np.pi / 2)  # Handle x=0 with a limit value
    return result

# Определяем функцию f для последовательности
def f(x):
    # Corrected exponentiation and ensured array-safe operations
    return (2 * x + 1) / (x + 2) + arccot(np.sqrt(2 + (-1) ** x))

# Генерируем массив натуральных чисел от 1 до 100
x = np.array(list(range(1, 101)))
y = f(x)

# Создаем график
plt.figure(figsize=(12, 6), facecolor='#0b1b3f')

ax = plt.gca()  # Получаем текущие оси
ax.set_facecolor('#020a26')  # Установка еще более темного синего фона

plt.scatter(x, y, color='#1f77b4', label='Значения последовательности', edgecolor='white', alpha=0.7)  # Цвет точек в синем оттенке
plt.plot(x, y, color='#1f77b4', label='Значения последовательности', alpha=0.7)  # Соединение точек линией синим цветом
plt.scatter(25, f(25), color='red', s=52, edgecolor='white', label='$x_{n}$ точка (для примера)')  # Подсветка точки с более светлым синим оттенком

# Горизонтальная линия для справки
plt.axhline(y=0, color='lightblue', linestyle='--', linewidth=0.8, alpha=0.6)  # Синяя горизонтальная линия

# Заголовки и метки осей
plt.title('Визуализация последовательности $x_{n}=\\frac{2n+1}{n+2}+arcctg\\sqrt{ 2+(-1)^n }$', fontsize=14, color='#add8e6')
plt.xlabel('Значения $n \in \mathbb{N}$', fontsize=15, color='#add8e6')
plt.ylabel('Значения $x_{n}$', fontsize=12, color='#add8e6')

# Добавление сетки
plt.grid(visible=True, which='both', color='lightblue', linestyle='--', linewidth=0.7, alpha=0.5)

# Показ значений на осях
plt.xticks(np.arange(0, 110, 10), color='#add8e6')  # Шаг значений по оси x в светлом синем
plt.yticks(np.linspace(min(y) - 2, max(y) + 3, 5), color='#add8e6')  # Шаг значений по оси y в светлом синем

# Легенда с изменением цвета текста
legend = plt.legend(facecolor='#0b1b3f')
for text in legend.get_texts():
    text.set_color('#add8e6')  # Установка цвета текста легенды на светло-синий

# Показ графика
plt.show()
```

4-5. По данному $\varepsilon>0$ определим номер $n_{0}$, начиная с которого, элементы подпоследовательности будут попадать в $\varepsilon \text{-окрестность}$ предела. Отедельно выделим те точки, которые попадают в данную $\varepsilon \text{-окрестность}$ и отметьте горизонтальной линией значение предела этой подпоследовательности.


1) Пусть $ɛ = 0.1$:

```python
# Импорт необходимых библиотек
import numpy as np
from matplotlib import pyplot as plt

# арккотангенс
def arccot(x):
    with np.errstate(divide='ignore', invalid='ignore'):
        result = np.where(x != 0, np.arctan(1 / x), np.pi / 2)  # Обработка случая x=0
    return result

# Определяем функцию f для последовательности
def f(x):
    return (2 * x + 1) / (x + 2) + arccot(np.sqrt(2 + (-1) ** x))

# Генерируем массив натуральных чисел от 1 до 100
x = np.array(list(range(1, 101)))
y = f(x)

# Подпоследовательность x_{2k}
x_even = x[1::2]  # Четные индексы (x_{2k}), начиная со второго элемента
y_even = f(x_even)

# Предел подпоследовательности
inf_xn = 2 + np.pi / 6
epsilon = 0.1

# Находим точки, которые попадают в epsilon-окрестность предела
indices_in_epsilon_neighborhood = np.abs(y_even - inf_xn) < epsilon
x_in_epsilon_neighborhood = x_even[indices_in_epsilon_neighborhood]
y_in_epsilon_neighborhood = y_even[indices_in_epsilon_neighborhood]

# Определяем n0 - первый индекс, начиная с которого элементы подпоследовательности находятся в epsilon-окрестности
n0 = x_in_epsilon_neighborhood[0] if len(x_in_epsilon_neighborhood) > 0 else None

# Создаем график
plt.figure(figsize=(12, 6), facecolor='#0b1b3f')

ax = plt.gca()  # Получаем текущие оси
ax.set_facecolor('#020a26')  # Установка темного синего фона

plt.scatter(x, y, color='#1f77b4', label='Значения последовательности $x_{n}$', edgecolor='white', alpha=0.7)  # Основные точки
plt.plot(x, y, color='#1f77b4', alpha=0.7)  # Соединение точек линией синим цветом
plt.scatter(x_even, y_even, color='purple', s=70, edgecolor='white', label='Подпоследовательность $x_{2k}$')  # Подпоследовательность
plt.scatter(x_in_epsilon_neighborhood, y_in_epsilon_neighborhood, color='red', s=100, edgecolor='white', label='Точки в $\\varepsilon$-окрестности предела')  # Точки в epsilon-окрестности

# Горизонтальная линия для предела
plt.axhline(y=inf_xn, color='cyan', linestyle='--', linewidth=1.5, label='$\\lim_{n \\to \\infty} x_{2k} = 2 + \\frac{\\pi}{6}$')

# Заголовки и метки осей
plt.title('Визуализация подпоследовательности $x_{2k}$ и $\\varepsilon$-окрестности предела при epsilon = 0.1', fontsize=14, color='#add8e6')
plt.xlabel('Значения $n \\in \\mathbb{N}$', fontsize=15, color='#add8e6')
plt.ylabel('Значения $x_{n}$', fontsize=12, color='#add8e6')

# Добавление сетки
plt.grid(visible=True, which='both', color='lightblue', linestyle='--', linewidth=0.7, alpha=0.5)

# Показ значений на осях
plt.xticks(np.arange(0, 110, 10), color='#add8e6')  # Шаг значений по оси x
plt.yticks(np.linspace(min(y) - 1, max(y) + 2, 5), color='#add8e6')  # Шаг значений по оси y

# Легенда с изменением цвета текста
legend = plt.legend(facecolor='#0b1b3f')
for text in legend.get_texts():
    text.set_color('#add8e6')  # Установка цвета текста легенды на светло-синий

# Показ графика
plt.show()

# Выводим n0
if n0:
    print(f"Начиная с n0 = {n0//2}, элементы подпоследовательности попадают в epsilon-окрестность предела.")
else:
    print("Нет элементов подпоследовательности, попадающих в указанную epsilon-окрестность.")
```

n0 = 15 — первый индекс, начиная с которого элементы подпоследовательности попадают в ε-окрестность.

```python
# Импорт необходимых библиотек
import numpy as np
from matplotlib import pyplot as plt

# арккотангенс
def arccot(x):
    with np.errstate(divide='ignore', invalid='ignore'):
        result = np.where(x != 0, np.arctan(1 / x), np.pi / 2)  # Обработка случая x=0
    return result

# Определяем функцию f для последовательности
def f(x):
    return (2 * x + 1) / (x + 2) + arccot(np.sqrt(2 + (-1) ** x))

# Генерируем массив натуральных чисел от 1 до 100
x = np.array(list(range(29941, 30031)))
y = f(x)

# Подпоследовательность x_{2k}
x_even = x[1::2]  # Четные индексы (x_{2k}), начиная со второго элемента
y_even = f(x_even)

# Определяем верхнюю границу (sup) и epsilon
sup_xn = 2 + np.pi / 6
epsilon = 0.0001  # Задаём значение ε

# Находим элементы, удовлетворяющие неравенству x_{m} > sup_xn - epsilon
indices_above_threshold = y_even > (sup_xn - epsilon)
x_above_threshold = x_even[indices_above_threshold]
y_above_threshold = y_even[indices_above_threshold]

# Создаем график
plt.figure(figsize=(12, 6), facecolor='#0b1b3f')

ax = plt.gca()  # Получаем текущие оси
ax.set_facecolor('#020a26')  # Установка темного синего фона

plt.scatter(x, y, color='#1f77b4', label='Значения последовательности $x_{n}$', edgecolor='white', alpha=0.7)  # Основные точки
plt.plot(x, y, color='#1f77b4', alpha=0.7)  # Соединение точек линией синим цветом
plt.scatter(x_even, y_even, color='purple', s=70, edgecolor='white', label='Подпоследовательность $x_{2k}$')  # Подпоследовательность
plt.scatter(x_above_threshold, y_above_threshold, color='orange', s=100, edgecolor='white', label='Точки $x_{m}$: $x_{m} > sup - \\varepsilon$')  # Точки выше границы

# Горизонтальная линия для верхней границы
plt.axhline(y=sup_xn, color='cyan', linestyle='--', linewidth=1.5, label='$sup\\text{ }x_{n} = 2 + \\frac{\\pi}{6}$')

# Заголовки и метки осей
plt.title('Визуализация подпоследовательности $x_{2k}$ и элементов выше $sup - \\varepsilon$', fontsize=14, color='#add8e6')
plt.xlabel('Значения $n \\in \\mathbb{N}$', fontsize=15, color='#add8e6')
plt.ylabel('Значения $x_{n}$', fontsize=12, color='#add8e6')

# Добавление сетки
plt.grid(visible=True, which='both', color='lightblue', linestyle='--', linewidth=0.7, alpha=0.5)

# Показ значений на осях
plt.xticks(np.arange(29941, 30031, 10), color='#add8e6')  # Шаг значений по оси x
plt.yticks(np.linspace(min(y) - 1, max(y) + 2, 5), color='#add8e6')  # Шаг значений по оси y

# Легенда с изменением цвета текста
legend = plt.legend(facecolor='#0b1b3f')
for text in legend.get_texts():
    text.set_color('#add8e6')  # Установка цвета текста легенды на светло-синий

# Показ графика
plt.show()

# Выводим количество элементов, удовлетворяющих условию
if len(x_above_threshold) > 0:
    print(f"Количество элементов, удовлетворяющих неравенству, на заданном промежутке: {len(x_above_threshold)}")
else:
    print("Нет элементов, удовлетворяющих заданному неравенству, на заданном промежутке.")
```
